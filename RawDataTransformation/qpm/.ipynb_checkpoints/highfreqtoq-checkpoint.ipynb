```python
df_dates = pd.DataFrame({'date': pd.date_range(start='1997-01-01', end='2019-09-30')})
```

```python
df_dates.head()
```

```python
df = pd.merge(df_dates, df, on='date', how='outer')
```

```python
quarterly_values = []
quarterly_dates = []
next_month = {1:4, 4:7, 7:10, 10:1}
for index, row in df.iterrows():
    if row['date'].month in {1,4,7,10} and row['date'].day == 1:
        print(f"{row['date']}: {row['value']}")
        found = False
        i = index+1
        while found == False:
            if df.loc[i, 'date'].month == next_month[row['date'].month] and df.loc[i, 'date'].day == 1:
                found = True
            else:
                i += 1
        print(f'{index}, {i}')
        quarterly_values.append(df.loc[index:i, 'value'].mean())
        quarterly_dates.append(row['date'])
            
```

```python
df2 = pd.DataFrame({'dates': quarterly_dates, 'values': quarterly_values})
```
